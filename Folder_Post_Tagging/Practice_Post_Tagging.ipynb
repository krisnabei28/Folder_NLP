```python
import pandas as pdData
from nltk import hmm
```

```python
dataset = pdData.read_table('Indonesian_Manually_Tagged_Corpus.tsv', header=None, names=['words','tags'])
counted = dataset['tags'].value_counts()

print(dataset)
print('Banyak Tags', counted.head(10), sep='\n')
```

```python
def listSentences(list_of_words, sep):
    sentence = []
    sentences = []
    for word in list_of_words:
        sentence.append(word)
        if word == sep:
            sentences.append(sentence)
            sentence = []
    return sentences
```

```python
list_of_words = dataset.to_records(index=False).tolist()
counter = ('.', 'Z')
```

```python
train_data = listSentences(list_of_words, counter)
trainer=hmm.HiddenMarkovModelTrainer()
```

```python
tagger=trainer.train_supervised(train_data)
loadData = open("/home/krisnabei/Documents/Dokumen_Krisna/NLP/Post_Tagging_Code/dataSet.txt", "r")
inputData = loadData.read()
```

```python
outputPOS = tagger.tag(inputData.split(' '))
print(outputPOS)
```
